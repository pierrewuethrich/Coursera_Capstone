# Part 2: Add longitude and latitude data to the dataframe

```python
postal_df=pd.read_csv('./postal_dataset.csv')
coord_df=pd.read_csv('/Users/pierrewuethrich/Downloads/Geospatial_Coordinates.csv')

#Remove empty row at index 0
postal_df=postal_df[1:]
postal_df.drop(columns=['Unnamed: 0'], inplace=True)
postal_df=postal_df.reset_index()
```

```python
longitude_vec = []
latitude_vec = []

for index in range(len(postal_df)):
    latitude = coord_df.loc[coord_df['Postal Code'] == postal_df['Postcode'][index]]['Latitude']
    longitude = coord_df.loc[coord_df['Postal Code'] == postal_df['Postcode'][index]]['Longitude']
    
    longitude_vec.append(np.array(longitude).squeeze())
    latitude_vec.append(np.array(latitude).squeeze())
```

```python
postal_df['Latitude']=latitude_vec
postal_df['Longitude']=longitude_vec
postal_df.drop(columns='index', inplace=True)
postal_df.head(25)
```
